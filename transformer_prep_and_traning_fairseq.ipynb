```
from fairseq.models.transformer import TransformerModel
model = TransformerModel.from_pretrained(
  model_name_or_path='.',
  checkpoint_file='checkpoints/checkpoint8.pt',
  data_name_or_path='../fairseq/',
  bpe= 'fastbpe',
  bpe_codes='bpe.32k/codes'
)
args = {'print_alignment':True,
        'tokenizer':'moses'}
input = model.encode('This is a test.')
hypos = model.generate(input, beam=5, verbose=False,**args)
for hypo in hypos:
  translation = model.decode(hypo['tokens'])
  print(translation)
  print(hypo['alignment'])
```

```
!fairseq-generate --path checkpoints/checkpoint8.pt ../fairseq/ --beam 5 --source-lang en --target-lang de --tokenizer moses --bpe fastbpe --bpe-codes bpe.32k/codes --remove-bpe | tee gen.out
```
